```python
cps = pd.read_csv('cps_00001.csv.gz', compression='gzip')
print(cps.shape)
print(cps.head())
print(cps.columns.tolist())
```

```python
# 취업자만 필터링 (EMPSTAT 10, 12 = 취업)
cps_employed = cps[cps['EMPSTAT'].isin([10, 12])].copy()

# OCC2010 코드 정리 (9999 = 미분류 제거)
cps_employed = cps_employed[cps_employed['OCC2010'] != 9999]

# NATIVITY 정리 (1 = 내국인, 4,5 = 이민자)
cps_employed = cps_employed[cps_employed['NATIVITY'].isin([1, 4, 5])]
cps_employed['immigrant'] = cps_employed['NATIVITY'].isin([4, 5]).astype(int)

print(cps_employed.shape)
print(cps_employed['immigrant'].value_counts())
```

```python
# OCC2010 코드를 문자열로 변환해서 SOC 형식에 맞게 만들기
# OCC2010은 숫자, SOC_6는 "11-1011" 형식

# 먼저 OCC2010 고유값 확인
print(cps_employed['OCC2010'].unique()[:20])
print(f"OCC2010 고유 직종 수: {cps_employed['OCC2010'].nunique()}")

# aioe_calc_6 SOC 형식 확인
print(aioe_calc_6['SOC_6'].head())
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import zipfile, gzip, shutil

# 1. Felten 데이터
A_j = pd.read_excel('AIOE_DataAppendix.xlsx', sheet_name='Appendix E')
A_j.columns = ['ability', 'A_j']

aioe_ref = pd.read_excel('AIOE_DataAppendix.xlsx', sheet_name='Appendix A')
aioe_ref.columns = ['SOC_code', 'occupation', 'AIOE']

# 2. O*NET
onet = pd.read_excel('Abilities.xlsx')
onet_im = onet[onet['Scale ID'] == 'IM'][['O*NET-SOC Code', 'Element Name', 'Data Value']].rename(columns={'Data Value': 'importance'})
onet_lv = onet[onet['Scale ID'] == 'LV'][['O*NET-SOC Code', 'Element Name', 'Data Value']].rename(columns={'Data Value': 'level'})
onet_merged = onet_im.merge(onet_lv, on=['O*NET-SOC Code', 'Element Name'])
onet_merged = onet_merged.merge(A_j, left_on='Element Name', right_on='ability', how='inner')

# 3. AIOE 계산
onet_merged['numerator'] = onet_merged['A_j'] * onet_merged['importance'] * onet_merged['level']
onet_merged['denominator'] = onet_merged['importance'] * onet_merged['level']
aioe_calc = onet_merged.groupby('O*NET-SOC Code').agg(
    numerator_sum=('numerator', 'sum'),
    denominator_sum=('denominator', 'sum')
).reset_index()
aioe_calc['AIOE_calc'] = aioe_calc['numerator_sum'] / aioe_calc['denominator_sum']
aioe_calc['SOC_6'] = aioe_calc['O*NET-SOC Code'].str[:7]
aioe_calc_6 = aioe_calc.groupby('SOC_6')['AIOE_calc'].mean().reset_index()
aioe_calc_6['AIOE_standardized'] = stats.zscore(aioe_calc_6['AIOE_calc'])

# 4. CPS
cps = pd.read_csv('cps_00001.csv.gz', compression='gzip')
cps_employed = cps[cps['EMPSTAT'].isin([10, 12])].copy()
cps_employed = cps_employed[cps_employed['OCC2010'] != 9999]
cps_employed = cps_employed[cps_employed['NATIVITY'].isin([1, 4, 5])]
cps_employed['immigrant'] = cps_employed['NATIVITY'].isin([4, 5]).astype(int)

print("완료")
print(f"AIOE 직종 수: {len(aioe_calc_6)}")
print(f"CPS 취업자 수: {len(cps_employed)}")
```

```python
# OCC2010 코드를 4자리 문자열로 변환 후 SOC 형식으로 맞추기
# OCC2010: 1010 → SOC: 11-1011 형식이라 직접 변환은 안 되고
# crosswalk 파일 사용

cw2 = pd.read_stata('occ2010_occ1990dd.dta')
cw2['occ'] = cw2['occ'].astype(str).str.zfill(4)

# OCC2010도 4자리로
cps_employed['OCC2010_str'] = cps_employed['OCC2010'].astype(str).str.zfill(4)

# CPS에 occ1990dd 붙이기
cps_employed = cps_employed.merge(cw2, left_on='OCC2010_str', right_on='occ', how='left')

print(f"merge 후 행 수: {len(cps_employed)}")
print(f"occ1990dd 매칭 안된 비율: {cps_employed['occ1990dd'].isna().mean():.2%}")
print(cps_employed[['OCC2010', 'occ1990dd']].head(10))
```

```python
# RTI 다시 계산
rti = pd.read_stata('occ1990dd_task_alm.dta')
rti['RTI'] = (np.log(rti['task_routine']) 
              - np.log(rti['task_abstract']) 
              - np.log(rti['task_manual']))
rti['RTI'] = rti['RTI'].replace([np.inf, -np.inf], np.nan)
rti = rti.dropna(subset=['RTI'])

# CPS에 RTI 붙이기
cps_employed = cps_employed.merge(rti[['occ1990dd', 'RTI']], on='occ1990dd', how='left')

# OCC2010 → SOC 변환 위해 crosswalk 사용
# occ2010 코드를 SOC 형식으로 변환
cw_soc = pd.read_excel('2019_to_SOC_Crosswalk.xlsx', header=2)
cw_soc.columns = ['onet_code', 'onet_title', 'soc_code', 'soc_title']
cw_soc = cw_soc.iloc[1:].reset_index(drop=True)
cw_soc = cw_soc.dropna(subset=['onet_code'])
cw_soc['SOC_6'] = cw_soc['soc_code'].str[:7]
cw_soc = cw_soc[['SOC_6']].drop_duplicates()

# AIOE를 SOC_6 기준으로 CPS에 붙이기
# OCC2010 숫자를 SOC 형식으로 변환
def occ2010_to_soc(occ):
    occ_str = str(int(occ)).zfill(4)
    return occ_str[:2] + '-' + occ_str[2:]

cps_employed['SOC_6'] = cps_employed['OCC2010'].apply(occ2010_to_soc)
cps_employed = cps_employed.merge(aioe_calc_6[['SOC_6', 'AIOE_standardized']], on='SOC_6', how='left')

print(f"AIOE 매칭률: {cps_employed['AIOE_standardized'].notna().mean():.2%}")
print(f"RTI 매칭률: {cps_employed['RTI'].notna().mean():.2%}")
print(cps_employed[['OCC2010', 'SOC_6', 'AIOE_standardized', 'RTI', 'immigrant']].head(10))
```

```python
# occ1990dd crosswalk를 통해 AIOE를 역으로 붙이기
# cw2: occ(OCC2010) → occ1990dd 있음
# aioe_calc_6: SOC_6 기준

# SOC_6를 OCC2010으로 변환하는 crosswalk 만들기
# onet_code(11-1011.00) → SOC_6(11-1011) → OCC2010

# 먼저 O*NET SOC를 4자리 숫자로 변환 시도
# SOC 11-1011 → 1110 (앞 2자리 + 뒤 2자리)
aioe_calc_6['occ_attempt'] = (aioe_calc_6['SOC_6'].str[:2] + 
                               aioe_calc_6['SOC_6'].str[3:5]).astype(int) * 10

print(aioe_calc_6[['SOC_6', 'occ_attempt']].head(10))
print(cw2['occ'].head(10))
```

```python
from rapidfuzz import process, fuzz

# OCC2010 코드-이름 딕셔너리 직접 만들기 (스크린샷에서 보이는 것들)
occ2010_names = {
    10: 'Chief executives and legislators',
    20: 'General and Operations Managers',
    40: 'Advertising and Promotions Managers',
    50: 'Marketing and Sales Managers',
    60: 'Public Relations and Fundraising Managers',
    100: 'Administrative Services Managers',
    110: 'Computer and Information Systems Managers',
    120: 'Financial Managers',
    135: 'Compensation and benefits managers',
    136: 'Human Resources Managers',
    137: 'Training and development managers',
    140: 'Industrial Production Managers',
    150: 'Purchasing Managers',
    160: 'Transportation, Storage, and Distribution Managers',
    205: 'Farmers, Ranchers, and Other Agricultural Managers',
    220: 'Construction Managers',
    230: 'Education Administrators',
    300: 'Architectural and Engineering Managers',
    310: 'Food Service Managers',
}

# aioe_ref 직종 이름 목록
aioe_names = aioe_ref['occupation'].tolist()
aioe_soc = aioe_ref['SOC_code'].tolist()

# fuzzy match
results = {}
for occ_code, occ_name in occ2010_names.items():
    match, score, idx = process.extractOne(occ_name, aioe_names, scorer=fuzz.token_sort_ratio)
    results[occ_code] = {'occ_name': occ_name, 'matched': match, 'SOC': aioe_soc[idx], 'score': score}

for k, v in list(results.items())[:10]:
    print(f"{k}: {v['occ_name']} → {v['matched']} ({v['SOC']}, {v['score']})")
```

```python
# RTI는 이미 CPS에 붙어있어 (매칭률 98.69%)
# AIOE는 occupation 이름 기반으로 CPS에 붙일 수 있어

# aioe_ref에 OCC2010 이름 컬럼 추가 (수동 매핑 대신)
# O*NET 데이터에 이미 SOC코드-직종명이 있으니까
# onet에서 SOC_6 기준으로 직종명 추출

onet_occ = onet[['O*NET-SOC Code', 'Title']].drop_duplicates()
onet_occ['SOC_6'] = onet_occ['O*NET-SOC Code'].str[:7]
onet_occ_6 = onet_occ.groupby('SOC_6')['Title'].first().reset_index()

# aioe_calc_6에 직종명 붙이기
aioe_with_name = aioe_calc_6.merge(onet_occ_6, on='SOC_6')
print(aioe_with_name.head())
print(aioe_with_name.shape)
```

```python
from rapidfuzz import process, fuzz

# IPUMS OCC2010 코드-이름 테이블 직접 만들기
# CPS에 있는 고유 OCC2010 코드 목록
unique_occ = cps_employed['OCC2010'].unique()
print(f"고유 OCC2010 코드 수: {len(unique_occ)}")

# aioe_with_name의 직종명 목록
aioe_titles = aioe_with_name['Title'].tolist()
aioe_soc = aioe_with_name['SOC_6'].tolist()
aioe_scores = aioe_with_name['AIOE_standardized'].tolist()

# OCC2010 코드별로 직종명이 필요한데
# cw2에 occ1990dd는 있고, aioe_ref에 SOC-직종명 있어
# occ1990dd → occ1990dd_task에 직종명 없음

# 대신: occ2010_occ1990dd crosswalk + aioe_ref 직종명으로
# occ1990dd 기준 AIOE 평균값 만들기

# Step 1: aioe_with_name에서 SOC_6 → AIOE
# Step 2: cw2에서 OCC2010 → occ1990dd
# Step 3: 둘을 occ1990dd 기준으로 연결하려면 SOC→occ1990dd 필요

# 현실적 대안: occ1990dd별 평균 AIOE를 
# Felten occupation 순서 기반으로 근사치 할당

# 지금 당장 가능한 분석: RTI만으로 이민자/내국인 비교
print("\n=== RTI 기반 이민자 vs 내국인 비교 ===")
rti_comparison = cps_employed.groupby('immigrant')['RTI'].mean()
print(rti_comparison)
```

```python
# OCC2010 코드별 직종명을 occ1990dd crosswalk 통해 간접적으로 얻기
# cw2: OCC2010 → occ1990dd
# rti df: occ1990dd 있음
# aioe_ref: SOC → occupation name

# 다른 방법: OCC2010 472개를 aioe 774개 직종명에 fuzzy match
# CPS OCC2010 고유값과 해당 직종 샘플 크기

occ_counts = cps_employed.groupby('OCC2010').size().reset_index(name='count')
occ_counts['OCC2010_str'] = occ_counts['OCC2010'].astype(str).str.zfill(4)

# aioe_with_name 직종명으로 fuzzy match
aioe_titles = aioe_with_name['Title'].tolist()

def match_occupation(occ_code):
    # occ1990dd 통해 간접 매칭 대신 코드 숫자 패턴으로 SOC 근사
    # OCC2010 0010 → SOC 11-1011 (Management 계열)
    # 코드 범위로 대분류 매칭
    code = int(occ_code)
    if code < 500: return '11'      # Management
    elif code < 800: return '13'    # Business/Financial
    elif code < 1000: return '15'   # Computer/Math
    elif code < 1300: return '17'   # Architecture/Engineering
    elif code < 1600: return '19'   # Life/Physical/Social Science
    elif code < 2000: return '21'   # Community/Social Service
    elif code < 2200: return '23'   # Legal
    elif code < 2600: return '25'   # Education
    elif code < 3000: return '27'   # Arts/Entertainment
    elif code < 3600: return '29'   # Healthcare Practitioners
    elif code < 4000: return '31'   # Healthcare Support
    elif code < 4200: return '33'   # Protective Service
    elif code < 4600: return '35'   # Food Preparation
    elif code < 5000: return '37'   # Building/Grounds
    elif code < 5500: return '39'   # Personal Care
    elif code < 6000: return '41'   # Sales
    elif code < 6600: return '43'   # Office/Admin
    elif code < 7000: return '45'   # Farming/Fishing
    elif code < 7700: return '47'   # Construction
    elif code < 8000: return '49'   # Installation/Maintenance
    elif code < 8900: return '51'   # Production
    else: return '53'               # Transportation

occ_counts['SOC_major'] = occ_counts['OCC2010_str'].apply(match_occupation)

# SOC 대분류별 평균 AIOE
aioe_with_name['SOC_major'] = aioe_with_name['SOC_6'].str[:2]
aioe_major = aioe_with_name.groupby('SOC_major')['AIOE_standardized'].mean().reset_index()
aioe_major.columns = ['SOC_major', 'AIOE_major']

print(aioe_major)
```

```python
# CPS에 SOC_major, AIOE_major 붙이기
cps_employed['OCC2010_str'] = cps_employed['OCC2010'].astype(str).str.zfill(4)
cps_employed['SOC_major'] = cps_employed['OCC2010_str'].apply(match_occupation)
cps_employed = cps_employed.merge(aioe_major, on='SOC_major', how='left')

# 교육 수준 분류
cps_employed['educ_group'] = pd.cut(
    cps_employed['EDUC'],
    bins=[0, 73, 111, 999],
    labels=['Less than HS', 'HS/Some College', 'College+']
)

# 핵심 분석 1: 전체 AIOE 비교
print("=== 전체 AIOE 비교 ===")
print(cps_employed.groupby('immigrant')['AIOE_major'].mean())

# 핵심 분석 2: 교육 수준별
print("\n=== 교육 수준별 AIOE 비교 ===")
print(cps_employed.groupby(['immigrant', 'educ_group'])['AIOE_major'].mean().unstack())

# 핵심 분석 3: 대분류별 이민자 비율
print("\n=== 대분류별 이민자 비율 ===")
sector = cps_employed.groupby('SOC_major').agg(
    immigrant_share=('immigrant', 'mean'),
    AIOE=('AIOE_major', 'first')
).reset_index()
print(sector.sort_values('AIOE', ascending=False))
```

```python
html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>AI Exposure and Immigrant Workers in the U.S.</title>
    <style>
        body { font-family: Georgia, serif; max-width: 860px; margin: 60px auto; padding: 0 20px; color: #222; line-height: 1.8; }
        h1 { font-size: 2em; margin-bottom: 0.2em; }
        h2 { font-size: 1.3em; margin-top: 2em; border-bottom: 1px solid #ddd; padding-bottom: 6px; }
        .meta { color: #888; font-size: 0.9em; margin-bottom: 2em; }
        .finding { background: #f9f9f9; border-left: 4px solid #333; padding: 12px 20px; margin: 20px 0; }
        img { width: 100%; margin: 30px 0; border: 1px solid #eee; }
        .limitation { background: #fff8f0; border-left: 4px solid #e8a000; padding: 12px 20px; margin: 20px 0; }
        footer { margin-top: 60px; font-size: 0.85em; color: #aaa; border-top: 1px solid #eee; padding-top: 20px; }
    </style>
</head>
<body>

<h1>AI Exposure and Immigrant Workers in the U.S.</h1>
<div class="meta">Minjun · July 2025 · Data: O*NET, Felten et al. (2021), IPUMS CPS 2017–2019</div>

<h2>Research Question</h2>
<p>Are immigrant workers in the United States more or less exposed to AI-driven occupational displacement than native workers — and does this vary by education level or sector?</p>
<p>This project does not claim to measure displacement directly. Instead, it asks a prior question: which group of workers is more likely to be in occupations where AI is most relevant? The answer shapes where displacement risk would concentrate <em>if</em> AI exposure translates into job loss.</p>

<h2>Background</h2>
<p>The task-based framework of Autor, Levy & Murnane (2003) argues that technology substitutes for routine, codifiable tasks while complementing nonroutine cognitive work. Peri & Sparber (2009) showed that immigrant and native workers sort into different task niches — immigrants concentrating in manual tasks, natives in communication-intensive roles. This project asks whether that sorting pattern means immigrants face systematically lower AI exposure than natives.</p>

<h2>Data & Method</h2>
<p>I replicated the AI Occupational Exposure (AIOE) index of Felten, Raj & Seamans (2021) using O*NET ability data. The index measures how much each occupation's task requirements overlap with current AI capabilities, across 774 occupations at the 6-digit SOC level. The replication achieved a correlation of 0.9949 with Felten et al.'s published values.</p>
<p>I then merged this index with IPUMS CPS microdata (2017–2019) covering 2.58 million employed workers, categorized by nativity status (native-born vs. foreign-born).</p>

<h2>Results</h2>
<img src="aioe_immigrant_analysis.png" alt="AI Exposure Analysis Charts">

<div class="finding">
    <strong>Finding 1:</strong> Immigrant workers have lower average AI exposure (AIOE = 0.011) than native workers (AIOE = 0.089). The gap is modest but consistent across all education levels.
</div>

<div class="finding">
    <strong>Finding 2:</strong> Even among college-educated workers, immigrants show lower AIOE (0.624) than natives (0.694) — suggesting the gap is not simply explained by education differences.
</div>

<div class="finding">
    <strong>Finding 3:</strong> Sectors with high AI exposure (Legal, IT, Business) have below-average immigrant shares. Sectors with high immigrant shares (Construction, Food, Farming) have the lowest AIOE scores.
</div>

<h2>Limitations</h2>
<div class="limitation">
    <strong>Exposure ≠ Displacement.</strong> AIOE measures occupational proximity to AI — not whether workers will actually be displaced. A high score means the job could be affected, not that it will be.
</div>
<div class="limitation">
    <strong>Sector-level approximation.</strong> Due to crosswalk limitations between OCC2010 and SOC codes, AIOE was assigned at the 2-digit sector level rather than the detailed occupation level. This understates within-sector variation.
</div>
<div class="limitation">
    <strong>Index vintage.</strong> The Felten et al. index was constructed before the large language model era (pre-ChatGPT). Post-2022 AI capabilities may shift exposure patterns significantly, particularly for cognitive tasks previously considered nonroutine.
</div>

<h2>What Comes Next</h2>
<p>This project establishes the exposure baseline. The natural follow-up is a causal design: did workers in high-AIOE occupations experience differential wage or employment changes after 2020, and does this effect differ between immigrant and native workers? That question requires panel data and a difference-in-differences design — the next project in this series.</p>

<h2>References</h2>
<p>Autor, D., Levy, F., & Murnane, R. (2003). The skill content of recent technological change. <em>Quarterly Journal of Economics</em>.<br>
Felten, E., Raj, M., & Seamans, R. (2021). Occupational, industry, and geographic exposure to artificial intelligence. <em>Strategic Management Journal</em>.<br>
Peri, G., & Sparber, C. (2009). Task specialization, immigration, and wages. <em>American Economic Journal: Applied Economics</em>.</p>

<footer>
    Part of an independent research portfolio on macroeconomic shocks and labor markets. 
    Code available on <a href="https://github.com/MinJTheNinja/ai-immigrant-labor">GitHub</a>.
</footer>

</body>
</html>"""

with open('index.html', 'w', encoding='utf-8') as f:
    f.write(html_content)

print("index.html 생성 완료")
```
